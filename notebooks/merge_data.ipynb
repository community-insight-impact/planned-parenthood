# STD
<i>Data source:</i> <br>
<i>Year:</i> 2019 <br>
<i>Notes:</i> Currently only includes MO.

```python
std = pd.read_csv(DATA_DIR + 'std_by_county_2019.csv')
std.shape
```

```python
std.head()
```

```python
std['County'] = [std['County'].iloc[i].capitalize() for i in range(std.shape[0])]
```

```python
std['Diagnosis'] = [std['Diagnosis'].iloc[i].capitalize() for i in range(std.shape[0])]
```

```python
std.head()
```

```python
# std_dummies = pd.get_dummies(std)
```

```python
std_ch = std[std['Diagnosis'] == 'Chlamydia']
std_go = std[std['Diagnosis'] == 'Gonorrhea']
std_sy = std[std['Diagnosis'] == 'Syphilis']
```

```python
std_ch.shape, std_go.shape, std_sy.shape
```

```python
std_ch.head()
```

```python
data.head()
```

```python
data = pd.merge(data, std_ch, on='County', how='left')
data = data.rename(columns={'count':'STD (Chlamydia)'})
data = data.drop('Diagnosis', axis=1)
check_shape(data)
```

```python
data = pd.merge(data, std_go, on='County', how='left')
data = data.rename(columns={'count':'STD (Gonorrhea)'})
data = data.drop('Diagnosis', axis=1)
check_shape(data)
```

```python
data = pd.merge(data, std_sy, on='County', how='left')
data = data.rename(columns={'count':'STD (Syphilis)'})
data = data.drop('Diagnosis', axis=1)
check_shape(data)
```
